# Gold medals per country

```python
country_gold_medals = df[df["Medal"]=='Gold'].groupby('NOC')['ID'].count()
country_gold_medals
```

```python
country_gold_medals.plot(kind='bar', figsize=(20, 8))
plt.title('Gold medals per country')
plt.xlabel('Sport')
plt.ylabel('No Of Gold Medal')
plt.xticks(rotation=90)
plt.show()
```

# Average Weight by Gender and Sport

```python
avg_weight_gender_sport=df.groupby(['Sport','Sex'])["Weight"].mean()
avg_weight_gender_sport
```

```python
avg_weight_gender_sport.plot(kind='bar', figsize=(30, 8))
plt.title('Average Weight by Gender and Sport')
plt.xlabel('Sport')
plt.ylabel('Average Weight')
plt.xticks(rotation=90)
plt.legend(title='Gender')
plt.show()
```

```python
year_participant_count =df.groupby("Year")["ID"].count()
year_participant_count.plot(kind="line",marker="o",figsize=(10,5))
plt.plot()
```

```python
country_avg_age=df.groupby("NOC")["Age"].mean().sort_values(ascending=False)
country_avg_age.head(10).plot(kind="bar")
plt.title("Top 10 Countries by AVG Age")
plt.xlabel("Countries")
plt.ylabel("AVG Age")
```

```python
sns.boxplot(x="Season",y="Age",data=df)
plt.title(" Distribution of  Age By Season")
plt.xlabel("Season")
plt.ylabel("Age")
```

```python
sns.violinplot(y="Height",x="Medal",data=df,palette="Set2")
plt.title(" Distribution of  Height By Medal")
plt.xlabel("Medal")
plt.ylabel("Age")
```

```python
sns.scatterplot(data=df,x="Height",y="Weight",hue="Medal",palette="Set1")
plt.title(" Height Vs Weight By Medal Status")
sns.set_style('whitegrid')
plt.xlabel("Height")
plt.ylabel("Weight")
```
